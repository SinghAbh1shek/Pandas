Inner join only merge common element/id

```python
merged_df = pd.merge(customers_data, purchases_data, on='id', how='inner')
merged_df
```

Left join merges the left side element and the common element

```python
merged_df_left = pd.merge(customers_data, purchases_data, on='id', how='left')
merged_df_left
```

Left join merges the right side element and the common element

```python
merged_df_right = pd.merge(customers_data, purchases_data, on='id', how='right')
merged_df_right
```

Imputation: With the measure of the central tandency [ mean, median, mode ]

How can we deal with the categorical data. -> Mode

```python
mode_gender = customers_data['gender'].mode()[0]  #we use [0] cuz incase if there is a balance data so it gives one that one result
mode_gender
```

```python
customers_data['gender'].fillna(mode_gender, inplace=True)    # Its place 'Male' on missing gender data
customers_data.isnull().sum()
```

How can we deal with the numeric data -> ffill(), befill(), mean, median

```python
customers_data['street_num'].fillna(method='ffill', inplace=True)   #ffill means forward fill
customers_data.isnull().sum()
```

```python
customers_data.head()
```

```python
customers_data['street_name'].fillna(method='ffill', inplace=True)
customers_data.isnull().sum()
```

```python
customers_data['street_suffix'].fillna(method='ffill', inplace=True)
customers_data.isnull().sum()
```

```python
customers_data.head(40)
```

```python
customers_data['city'].fillna(method='bfill', inplace=True)
customers_data['state'].fillna(method='bfill', inplace=True)
customers_data['postcode'].fillna(method='bfill', inplace=True)
customers_data.isnull().sum()
```
