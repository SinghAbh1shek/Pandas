```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('/content/purchases_lyst1748071489995.xlsx')
data
```

```python
data.drop(columns='Unnamed: 0', inplace=True)
```

```python
data.head()
```

convert purch_date->datetime

```python
data['purch_date'] = pd.to_datetime(data['purch_date'])
data.dtypes
```

convert id-> object

```python
data['id'] = data['id'].astype(object)
data.dtypes
```

convert paid-> float

```python
data['paid'] = data['paid'].astype(float) #we can't convert that its contain string/object
data.dtypes
```

```python
# paid -> data preprocess and then only we can do the type conversion

data['paid'] = data['paid'].str.replace('[\$,]', '', regex=True)    #Its remove the $ and , form data['paid']
data.head()
```
