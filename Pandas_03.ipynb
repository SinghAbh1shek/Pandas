Access the rows in a given dataframe:.loc(), .iloc()

```python
dataframe = {
    'name': ['Arun', 'Amit', 'Anuj', 'Anup', 'Aman'],
    'Age':[22, 20, 24, 31, 21],
    'city': ['Jaipur', 'Bhopal', 'Patna', 'Punjab', 'Siwan']
}

dataframe = pd.DataFrame(dataframe, index = ['a', 'b', 'c', 'd', 'e'])
dataframe
```

.loc only works on customized index
and the last element is inclusive

```python
dataframe.loc['b']
```

```python
dataframe.loc['b':'d']
```

.iloc only works on default indexing
and the last element is exclusive while slicing

```python
dataframe.iloc[1:4]
```

```python
dataframe_v1 = {
    'name': ['Arun', 'Amit', 'Anuj', 'Anup', 'Aman'],
    'Age':[22, 20, 24, 31, 21],
    'city': ['Jaipur', 'Bhopal', 'Patna', 'Punjab', 'Siwan']
}

dataframe_v1 = pd.DataFrame(dataframe_v1)
dataframe_v1
```

```python
dataframe_v1.loc[3]   # In this case loc works same as iloc cuz here default and custimized indexing are same
```
